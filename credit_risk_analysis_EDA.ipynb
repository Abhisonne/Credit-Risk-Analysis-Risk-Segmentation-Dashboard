```python
import pandas as pd
df = pd.read_csv('/content/german_credit_data_updated.csv')
df.head()
```

```python
df.info()
print(df.describe())
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.columns
```

```python
df['Saving accounts'].fillna('No Saving', inplace=True)
df['Checking account'].fillna('No Checking', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Credit Risk'].value_counts()
```

```python
pd.crosstab(df['Checking account'], df['Credit Risk'])
```

```python
df.groupby('Credit Risk')['Credit amount'].mean()
```

```python
df.groupby('Credit Risk')['Duration'].mean()
```

```python
df.groupby('Credit Risk')['Age'].mean()
```

```python
pd.crosstab(df['Housing'], df['Credit Risk'])
```
